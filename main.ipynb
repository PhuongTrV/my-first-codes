```python
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv("data.csv")
df=df.rename({"TRƯỜNG THPT QUỲNH LƯU 3":"Stt", "Unnamed: 1":"Id", "Unnamed: 2":"Class", "Unnamed: 3":"Name", "Unnamed: 4":"Birthday", "Unnamed: 5":"Gender"},axis=1)
df = df[pd.notnull(df['Gender'])]
lst=[[0]]
i=0
for index,row in df.iterrows():
    lst2nd=list(row)
    if lst2nd[0]=="STT":
        i+=1
        lst.append([])
    if lst2nd[2]=="10A1":
        lst[i].append(lst2nd[3])
xl=[str(x) for x in range(0,len(lst))]
del xl[0]
yl=[]
for j in range(1,len(lst)):
    yl.append(len(lst[j]))
plt.plot(xl,yl)
plt.xlabel("Phòng")
plt.ylabel("Số học sinh 10a1")
```
